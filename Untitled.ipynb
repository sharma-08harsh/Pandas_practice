#### Creating a Dataframe Using Dictionary

```python
data = {'Name':['Harshit', 'Tom', 'Krish'], 'Age':[21, 22, 26]}
pd.DataFrame(data)
```

```python
data2 = {'Name'          :   ['tom' , 'chad' , 'sam'],
         'Age'           :   [23,24,25],
         'ID'            :   [1,2,3],
         'Qualification' :   ['bca', 'mba', 'bsc']}
df = pd.DataFrame(data2)
df
```

### Slicing in Dataframe Using Iloc and LOC

```python
data = {'one'    : pd.Series([1,2,3,4]),
        'two'    : pd.Series([10,20,30,40]),
        'three'  : pd.Series([100,200,300,400]),
        'four'   : pd.Series([1000,2000,3000,400])}
df = pd.DataFrame(data)
df.loc[:]
```

```python
df.loc[1:2, 'two':'three']
```

```python
df.iloc[ 0:4 , 0:4 ]
```

Selecting specific rows

```python
df.iloc[[0,2],[1,3]]
```

### Slicing using Conditions

```python
df.loc[df['two']>20,['three']]
```

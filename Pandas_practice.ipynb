### Slicing in Dataframe Using Iloc and LOC

```python
data = {'one'    : pd.Series([1,2,3,4]),
        'two'    : pd.Series([10,20,30,40]),
        'three'  : pd.Series([100,200,300,400]),
        'four'   : pd.Series([1000,2000,3000,400])}
df = pd.DataFrame(data)
df.loc[:]
```

```python
df.loc[1:2, 'two':'three']
```

```python
df.iloc[ 0:4 , 0:4 ]
```

Selecting specific rows

```python
df.iloc[[0,2],[1,3]]
```

### Slicing using Conditions

```python
df.loc[df['two']>20,['three']]
```

### Pipe Funtion

```python
def add_(i,j):
    return i+j
def sub_(i,j):
    return i-j
```

```python
df.pipe(add_,10)
```

```python
df.pipe(sub_,10)
```

```python
def mean_(col):
    return col.mean()

def square_(i):
    return i**2

print(df.pipe(mean_))
df.pipe(mean_).pipe(square_)
```
